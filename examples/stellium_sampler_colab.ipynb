```python
from IPython.display import SVG, display


def show_svg(filename):
    display(SVG(filename=filename))
```

## 2-Line QuickStart

```python
from stellium import ChartBuilder

chart = ChartBuilder.from_notable("Albert Einstein").with_aspects().calculate()
chart.draw("einstein.svg").save()

show_svg("einstein.svg")
```

## Custom Chart Visuals

```python
# Apply a preset for instant results
chart.draw("detailed.svg").preset_detailed().save()

# Choose a theme
chart.draw("midnight.svg").with_theme("midnight").save()

# Full customization
chart.draw("custom.svg").with_theme("celestial").with_zodiac_palette(
    "rainbow_celestial"
).with_moon_phase(position="bottom-left", show_label=True).with_chart_info(
    position="top-left"
).save()

show_svg("custom.svg")
```

## Feature: Zodiacal Releasing (L1-L4)

```python
from stellium.engines.releasing import ZodiacalReleasingAnalyzer

part = "Part of Fortune"

chart = (
    ChartBuilder.from_notable("Albert Einstein")
    .add_analyzer(ZodiacalReleasingAnalyzer([part]))
    .calculate()
)

zr = chart.zodiacal_releasing(part)
timeline = zr.l1_periods()

print(f"Zodiacal Releasing ({part}) L1 Periods:")
print("-" * 50)
for period in timeline:
    print(
        f"{period.start.date()} to {period.end.date()}: {period.sign} (Quality: {period.sentiment})"
    )
```

## Try Your Own Chart!

```python
my_chart = ChartBuilder.from_details(
    "1990-01-01 12:00", "London, UK", name="Me"
).calculate()

my_chart.draw("me.svg").with_theme("midnight").save()
show_svg("me.svg")
```
